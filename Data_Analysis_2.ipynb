# Wohnungssituation in Berlin und Hamburg: Vergleich

# Übersicht Datensätze

```python
# Load the packages
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
# Load the data
berlin_df = pd.read_csv('./df_berlin.csv')
hamburg_df = pd.read_csv('./df_hamburg.csv')
```

```python
berlin_df.info()
```

```python
hamburg_df.info()
```

Im Sommer 2022 standen mehr Wohnung in Berlin (80 Wohnungen) als in Hamburg (20 Wohnungen) zur Verfügung. 

# Berlin

```python
berlin_df_sum = berlin_df.groupby('district')['price (in EUR)'].describe().round()
berlin_df_sum
```

Der Datensatz enthält einige Ausreißer (z.B. die Wohnung aus Rudow). Grund darin könnte die Webseite selbst zu sein, die zusammen mit den Mietwohnungen als Werbung Kaufwohnungen vorschlägt. Deswegen werden die Wohnungen mit viel zu hohen Preisen aus dem Datensatz ausgeschlossen.

```python
# Ausschluss von Ausreißen
rent_be = berlin_df[berlin_df['price (in EUR)'] <= 10000]
rent_be.rename(columns = {'price (in EUR)':'price', 'size (in m^2)':'size'})
rent_be_sum = rent_be.groupby('district')['price (in EUR)'].describe().reset_index()
rent_be_sum_sorted = rent_be_sum.sort_values('mean', ascending = False)
```

```python
# Graphische Darstellung
plot_order = rent_be.groupby('district')['price (in EUR)'].mean().sort_values(ascending = False).index.values

sns.set_color_codes("pastel")
g = sns.catplot(data = rent_be, y = 'district', x = 'price (in EUR)', kind = 'bar',
               sharex = False, sharey = False, color = 'b', errwidth = 0, orient = 'h',
               order = plot_order)
g.set_axis_labels('Price (in EUR)','')
g.figure.set_size_inches(15, 12)
g.fig.tight_layout()
g.set(title = 'Übersicht von verfügbaren Mietwohnungen in Berlin pro Bezirk', xlim = (0, 4000))
plt.show()
```

```python
rent_be_sum_sorted.describe()
```

## Wie viel kostet Analog unserer Wohnung in anderen Bezirken?

Die erste Frage ist, wie viel Wohnungen sind analog zu unserer: ca. 90 m^2, 3 Zimmer, mit Balkon

```python
analog_be = rent_be[(rent_be['size (in m^2)'] >= 85) 
                    & (rent_be.balcony == 1) 
                    & (rent_be['size (in m^2)'] <= 95) 
                    & (rent_be.built_in_kitchen == 1)]
print('Insgesamt gibt es ', len(analog_be), ' Wohnungen in Berlin, die analog zur heutigen Wohnung vermietet wurden.')
```

In welchen Bezirke befinden sich die Wohnungen?

```python
analog_be
```

```python
ideal_be = rent_be[(rent_be.balcony == 1) 
                    & (rent_be['size (in m^2)'] >= 90) 
                    & (rent_be.built_in_kitchen == 1)
                    & (rent_be.basement == 1)]
print('Insgesamt gibt es ', len(ideal_be), ' Wohnungen in Berlin, die unseren Ansprüchen entsprechen.')
```

# Hamburg

```python
hamburg_df_sum = hamburg_df.groupby('district')['price (in EUR)'].describe().round()
hamburg_df_sum
```

Wie bei den Berliner Wohnungen, enthält der Datensatz aus Hamburg auch ein paar Ausreißer. Da die Quelle für beide Datensätze dieselbe ist, kann es wiederum an die Werbung von Kaufwohnungen liegen. Dementsprechend werden diese Ausreißer ausgeschlossen.
